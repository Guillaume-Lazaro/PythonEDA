```python
# 1) Import
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# 2) Infos
house_pricing = pd.read_csv('data/house_pricing.csv')
house_pricing.info()
```

```python
house_pricing
```

```python
# 3)
ax = sns.relplot(x='LotArea', y='SalePrice', data=house_pricing, color='red')
#ax.set(xlabel='LotArea', ylabel='SalePrice')
plt.show()
```

```python
# 4)
df_housing_prices = pd.DataFrame(house_pricing)
df_cleaned0 = df_housing_prices[ df_housing_prices['LotArea'] < 20000 ]
df_cleaned1 = df_cleaned0[ df_cleaned0['SalePrice'] < 500000 ]
df_cleaned1
```

```python
ax = sns.relplot(x='LotArea', y='SalePrice', data=df_cleaned1, color='red')
```

```python
# 5)
frontage_area = sns.relplot(x='LotArea', y='LotFrontage', data=df_cleaned1)
```

```python
# 6)
df_cleaned2 = df_cleaned1[ df_cleaned1['LotArea'] < 100000 ]
df_cleaned3 = df_cleaned2[ df_cleaned2['LotFrontage'] < 200 ]
frontage_area = sns.relplot(x='LotArea', y='LotFrontage', data=df_cleaned3)
```
